# Amprius Company & Stock Analysis

```python
import yfinance as yf

ampx_tick = 'AMPX'
ampx_data = yf.Ticker(ampx_tick)
```

```python
%matplotlib tk
```

```python
# Options:
# period: '1d', '5d', '1mo', '3mo', '6mo', '1y', '5y', 'max'
# interval: '1m', '5m', '15m', '1h', '1d', '1wk', '1mo'

period_str = "max"
interval_str = "1d"
ampx_hist = ampx_data.history(period=period_str, interval=interval_str)
ampx_ma_20_day = ampx_hist["Close"].rolling(window=20, min_periods=1).mean()
ampx_ma_50_day = ampx_hist["Close"].rolling(window=50, min_periods=1).mean()
ampx_ma_120_day = ampx_hist["Close"].rolling(window=120, min_periods=1).mean()

# Key Event Dates
import datetime
liberation_day = datetime.date(2025, 4, 2)
big_beautfiul_bill_signing = datetime.date(2025, 7, 4)

Q3_2022_earnings = datetime.date(2022, 11, 2)
Q4_2022_earnings = datetime.date(2023, 3, 23)

Q1_2023_earnings = datetime.date(2023, 5, 10)
Q2_2023_earnings = datetime.date(2023, 8, 4)
Q3_2023_earnings = datetime.date(2023, 11, 9)
Q4_2023_earnings = datetime.date(2024, 3, 21)

Q1_2024_earnings = datetime.date(2024, 5, 9)
Q2_2024_earnings = datetime.date(2024, 8, 8)
Q3_2024_earnings = datetime.date(2024, 11, 7)
Q4_2024_earnings = datetime.date(2025, 3, 20)

Q1_2025_earnings = datetime.date(2025, 5, 8)
Q2_2025_earnings = datetime.date(2025, 8, 7)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(ampx_hist.index, ampx_hist["Close"], label = 'Close Price', color = "blue")
plt.plot(ampx_hist.index, ampx_ma_20_day, label = '20-Day MA', color = "orange", alpha = 0.5)
plt.plot(ampx_hist.index, ampx_ma_50_day, label = '50-Day MA', color = "red", alpha = 0.5)
plt.plot(ampx_hist.index, ampx_ma_120_day, label = '120-Day MA', color = "green", alpha = 0.5)
plt.axvline(liberation_day, label='Liberation Day', color = 'black', linestyle = '-.')
plt.axvline(big_beautfiul_bill_signing, label='Big Beautiful Bill', color =  "orchid", linestyle = '-.')
plt.axvline(Q3_2022_earnings, label='Q3 2022 Earn.', color = 'pink', linestyle = '-.')
plt.axvline(Q4_2022_earnings, label='Q4 2022 Earn.', color = 'gray', linestyle = '-.')
plt.axvline(Q1_2023_earnings, label='Q1 2023 Earn.', color = 'magenta', linestyle = '-.')
plt.axvline(Q2_2023_earnings, label='Q2 2023 Earn.', color = 'yellow', linestyle = '-.')
plt.axvline(Q3_2023_earnings, label='Q3 2023 Earn.', color = 'magenta', linestyle = '-.')
plt.axvline(Q4_2023_earnings, label='Q4 2023 Earn.', color = 'chartreuse', linestyle = '-.')
plt.axvline(Q1_2024_earnings, label='Q1 2024 Earn.', color = 'cyan', linestyle = '-.')
plt.axvline(Q2_2024_earnings, label='Q2 2024 Earn.', color = 'green', linestyle = '-.')
plt.axvline(Q3_2024_earnings, label='Q3 2024 Earn.', color = 'orange', linestyle = '-.')
plt.axvline(Q4_2024_earnings, label='Q4 2024 Earn.', color = 'red', linestyle = '-.')
plt.axvline(Q1_2025_earnings, label='Q1 2025 Earn.', color = "lime", linestyle = '-.')
plt.axvline(Q2_2025_earnings, label='Q2 2025 Earn.', color = 'brown', linestyle = '-.')
plt.title("Amprius Close Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1),
                         loc='upper left', borderaxespad=0.)
plt.tight_layout()
plt.tick_params("x", rotation = 45)
plt.xlim([ampx_hist.index[0], ampx_hist.index[-1]])
plt.show()
```
